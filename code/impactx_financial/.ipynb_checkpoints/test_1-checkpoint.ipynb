```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import xgboost as xgb
from sklearn.metrics import classification_report, roc_auc_score
import shap
```

```python
data = pd.read_csv("../../dataset/creditcard.csv")
#Ê£ÄÊü•Âü∫Êú¨‰ø°ÊÅØ
print(data.head(10))
print(data.info())
print(data['Class'].value_counts())
```

ÊÄªÁöÑÊ†∑Êú¨Êï∞ÊòØ284807ÔºåÁ±ªÂûã0Êúâ284315ÔºõÁ±ªÂûã1Êúâ492.

```python
# ÊèêÂèñÊ†áÁ≠æ
y = data['Class'].values

# ÊèêÂèñÁâπÂæÅ
X = data.drop(columns=['Class']).values
fraud_ratio = sum(y) / len(y)#Âõ†‰∏∫ÊòØÔºà0Ôºå1ÔºâÊ†áÁ≠æÔºåÊ±ÇÂíåÂ∞±ÊòØ1Á±ªÊ†áÁ≠æÁöÑÊï∞„ÄÇ
print(f"Ê¨∫ËØàÂç†ÊØî: {fraud_ratio * 100:.4f}%")
```

```python
#Ê†áÂáÜÂåñ
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"ËÆ≠ÁªÉÈõÜÊ†∑Êú¨Êï∞: {len(X_train)}, ÊµãËØïÈõÜÊ†∑Êú¨Êï∞: {len(X_test)}")
print(f"ËÆ≠ÁªÉÈõÜ‰∏≠Ê¨∫ËØàÊØî‰æã: {sum(y_train)/len(y_train):.4f}")
```

**‰ΩøÁî®SOOMTEÊñπÊ≥ï**
Êàë‰ª¨‰ΩøÁî® imblearn Â∫ì‰∏≠ÁöÑ SMOTEÔºö
pip install imbalanced-learn


```python
# ÂàùÂßãÂåñ SMOTEÔºà‰ΩøÁî®ÈªòËÆ§ k=5Ôºâ
smote = SMOTE(random_state=42)

# ‰ªÖÂØπËÆ≠ÁªÉÈõÜËøõË°åËøáÈááÊ†∑
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print(f"ËøáÈááÊ†∑ÂêéËÆ≠ÁªÉÊ†∑Êú¨Êï∞: {len(X_train_resampled)}")
print(f"Ê≠£Á±ªÂç†ÊØî: {sum(y_train_resampled)/len(y_train_resampled):.4f}")
```

```python
pip install shap
```

**MÊ®°ÂùóÔºåÁâπÂæÅÊèêÂèñ**
ËÆ≠ÁªÉ XGBoost Ê®°ÂûãÔºà‰Ωú‰∏∫ IMPACTX ‰∏≠ÁöÑÁâπÂæÅÊèêÂèñÂô® MÔºâÂπ∂‰øùÂ≠òÊ®°Âûã„ÄÇ

```python
# ÂàùÂßãÂåñ XGBoost ÂàÜÁ±ªÂô®
model_xgb = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=1,  # Â∑≤‰ΩøÁî®SMOTEÔºå‰∏çÂÜçÈúÄË¶ÅÊùÉÈáçË∞ÉËäÇ
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# ËÆ≠ÁªÉÊ®°Âûã
model_xgb.fit(X_train_resampled, y_train_resampled)

# Âú®ÊµãËØïÈõÜ‰∏äËØÑ‰º∞
y_pred_proba = model_xgb.predict_proba(X_test)[:, 1]
y_pred_label = (y_pred_proba > 0.5).astype(int)

# ËæìÂá∫ÊÄßËÉΩÊåáÊ†á
print("Test Classification Report:")
print(classification_report(y_test, y_pred_label))
print(f"AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
```

```python
# ‰øùÂ≠òÊ®°ÂûãÂà∞Êñá‰ª∂
model_xgb.save_model("xgb_feature_extractor.json")
```

```python

# ÈáçÊñ∞Âä†ËΩΩÊ®°ÂûãÔºàÂèØÈÄâÔºâ
model_xgb = xgb.XGBClassifier()
model_xgb.load_model("xgb_feature_extractor.json")#Âä†ËΩΩËÆ≠ÁªÉÂ•ΩÁöÑÊ®°Âûã
```

```python
# ‰ΩøÁî® TreeExplainer
explainer = shap.TreeExplainer(model_xgb)
```

```python
# ÁîüÊàêËß£ÈáäÂÄºÔºàÁª¥Â∫¶: [n_samples, n_features]Ôºâ
shap_values = explainer.shap_values(X_train_resampled)

# shap_values ÊòØ numpy Êï∞ÁªÑÔºåÁõ¥Êé•‰Ωú‰∏∫ r_train
r_train = shap_values
print(f"Ëß£ÈáäÂõæ r_train ÂΩ¢Áä∂: {r_train.shape}")
```

```python
# ‰øùÂ≠ò SHAP ËæìÂá∫‰∏∫ .npy Êñá‰ª∂
np.save("r_train.npy", r_train)

# ÂêåÊó∂‰øùÂ≠òËÆ≠ÁªÉÊ†∑Êú¨ÂíåÊ†áÁ≠æÔºàÁ°Æ‰øùÂØπÈΩêÔºâ
np.save("X_train_resampled.npy", X_train_resampled)
np.save("y_train_resampled.npy", y_train_resampled)
```
