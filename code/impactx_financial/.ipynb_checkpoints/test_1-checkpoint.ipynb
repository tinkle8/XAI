```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
import xgboost as xgb
from sklearn.metrics import classification_report, roc_auc_score
import shap
```

```python
data = pd.read_csv("../../dataset/creditcard.csv")
#检查基本信息
print(data.head(10))
print(data.info())
print(data['Class'].value_counts())
```

总的样本数是284807，类型0有284315；类型1有492.

```python
# 提取标签
y = data['Class'].values

# 提取特征
X = data.drop(columns=['Class']).values
fraud_ratio = sum(y) / len(y)#因为是（0，1）标签，求和就是1类标签的数。
print(f"欺诈占比: {fraud_ratio * 100:.4f}%")
```

```python
#标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"训练集样本数: {len(X_train)}, 测试集样本数: {len(X_test)}")
print(f"训练集中欺诈比例: {sum(y_train)/len(y_train):.4f}")
```

**使用SOOMTE方法**
我们使用 imblearn 库中的 SMOTE：
pip install imbalanced-learn


```python
# 初始化 SMOTE（使用默认 k=5）
smote = SMOTE(random_state=42)

# 仅对训练集进行过采样
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print(f"过采样后训练样本数: {len(X_train_resampled)}")
print(f"正类占比: {sum(y_train_resampled)/len(y_train_resampled):.4f}")
```

```python
pip install shap
```

**M模块，特征提取**
训练 XGBoost 模型（作为 IMPACTX 中的特征提取器 M）并保存模型。

```python
# 初始化 XGBoost 分类器
model_xgb = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=1,  # 已使用SMOTE，不再需要权重调节
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# 训练模型
model_xgb.fit(X_train_resampled, y_train_resampled)

# 在测试集上评估
y_pred_proba = model_xgb.predict_proba(X_test)[:, 1]
y_pred_label = (y_pred_proba > 0.5).astype(int)

# 输出性能指标
print("Test Classification Report:")
print(classification_report(y_test, y_pred_label))
print(f"AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
```

```python
# 保存模型到文件
model_xgb.save_model("xgb_feature_extractor.json")
```

```python

# 重新加载模型（可选）
model_xgb = xgb.XGBClassifier()
model_xgb.load_model("xgb_feature_extractor.json")#加载训练好的模型
```

```python
# 使用 TreeExplainer
explainer = shap.TreeExplainer(model_xgb)
```

```python
# 生成解释值（维度: [n_samples, n_features]）
shap_values = explainer.shap_values(X_train_resampled)

# shap_values 是 numpy 数组，直接作为 r_train
r_train = shap_values
print(f"解释图 r_train 形状: {r_train.shape}")
```

[样本数, 特征数]= [454902, 30]
是SMOTE处理后的训练集样本数量，表示我们从原始训练集（约22万条）过采样生成了平衡后的数据。🤔生成的数据是否合规？
30，是每条交易记录的特征数量（包括Time，Amount，V1-V28，共30个特征）
•r_train[i] 表示第 i 条交易的解释图（一个长度为 30 的向量）
•r_train[i][j] 表示 SHAP 对该交易样本在第 j 个特征上的归因值（即该特征对欺诈与否的正负影响）



```python
# 保存 SHAP 输出为 .npy 文件
np.save("r_train.npy", r_train)

# 同时保存训练样本和标签（确保对齐）
np.save("X_train_resampled.npy", X_train_resampled)
np.save("y_train_resampled.npy", y_train_resampled)
```

```python
# 可视化前5个样本的 SHAP 图
shap.summary_plot(shap_values[:500], features=X_train_resampled[:500], feature_names=data.columns[:-1])
```

图表列出了很多特征，比如V1到V27，还有Time。SHAP值从-3到2，不是全部特征都有较大的影响。比如V14、V12、V8这些在顶部，SHAP值绝对值大，说明对模型影响大。而像V17、V26这些在中间，影响较小。

接下来，我应该解释怎么分析这张图。比如，V14的SHAP值分布在-3到2之间，颜色变化多，说明不同样本中V14的值对预测的影响差异大。反观V17，SHAP值接近0，说明影响不大。
